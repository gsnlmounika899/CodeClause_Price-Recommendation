# Price Histogram

```python
Fig,Ax=plt.subplots(figsize=(20,7))
Ax.hist(Price_Recommend['price'],bins=25,range=[0,1500],label="PRICE")
plt.title("PRODUCT PRICES",fontsize=17)
plt.xlabel('Price',fontsize=14)
plt.ylabel('Occurences',fontsize=14)
plt.grid()
plt.show()
```

# BoxPlot

```python
Fig,Ax=plt.subplots(figsize=(20,7))
sns.boxplot(x='price',data=Price_Recommend,showfliers=False)
plt.title("PRODUCT PRICES",fontsize=15)
plt.xlabel('Price',fontsize=14)
plt.show()
```

```python
Fig,Ax=plt.subplots(figsize=(20,5))
Trade_Names=Price_Recommend['brand_name'].value_counts()[0:10]
sns.barplot(x=Trade_Names.values,y=Trade_Names.index)
plt.title("TOP 10 BRANDS",fontsize=15)
plt.xlabel('Occurence',fontsize=14)
plt.ylabel('Brand_Names',fontsize=14)
plt.show()
```

# SHIPPING HISTOGRAM

```python
fig, ax = plt.subplots(figsize=(20, 7))
ax.hist(Price_Recommend['shipping'], range=[0, 1], rwidth=0.8, label="PRICE")
ax.set_title("SHIPPING", fontsize=17)
ax.set_xlabel('Shipping', fontsize=14)
ax.set_ylabel('Occurrences', fontsize=14)
ax.grid()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 7))
ax.hist(Price_Recommend.item_condition_id, range=[1, 5], rwidth=0.5, label="PRICE")
ax.set_title("ITEM CONDITION ID", fontsize=17)
ax.set_xlabel('item_condition_id', fontsize=14)
ax.set_ylabel('Occurrences', fontsize=14)
ax.grid()
plt.show()
```

# Price vs Shipping

```python
fig, ax = plt.subplots(figsize=(20, 7))
sns.boxplot(x=Price_Recommend['shipping'], y=Price_Recommend['price'], order=[0, 1])
ax.set_ylim(0, 70)
plt.title("PRICE & SHIPPING", fontsize=17)
plt.xlabel('Shipping', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 7))
sns.boxplot(x=Price_Recommend.item_condition_id, y=Price_Recommend['price'], order=[1,2,3,4, 5])
ax.set_ylim(0, 80)
plt.xlabel('item_condition_id', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.show()
```

```python
categories = Price_Recommend['category_name'].str.split("/", expand=True)
Category_1 = categories[0]
fig, ax = plt.subplots(figsize=(20, 7))
sns.boxplot(x=Category_1, y=Price_Recommend['price'])
ax.set_ylim(0, 80)
plt.xlabel('category_1', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.show()
```
